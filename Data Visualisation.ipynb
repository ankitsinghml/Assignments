```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_data = iris.data
iris_label = iris.target

fig = plt.figure(1, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=150)

pca = decomposition.PCA()
pca.n_components = 3
pca_data = pca.fit_transform(iris_data)

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        pca_data[iris_label == label, 0].mean(),
        pca_data[iris_label == label, 1].mean() + 1.5,
        pca_data[iris_label == label, 2].mean(),
        name,
        horizontalalignment="center",)

ax.scatter(pca_data[:,0], pca_data[:,1], pca_data[:,2], c=iris_label, cmap=plt.cm.viridis_r, edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
